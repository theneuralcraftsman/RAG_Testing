```python
import tiktoken

tokenizer = tiktoken.get_encoding('cl100k_base')

# create the length function
def tiktoken_len(text):
    tokens = tokenizer.encode(
        text,
        disallowed_special=()
    )
    return len(tokens)

tiktoken_len("hello I am a chunk of text and using the tiktoken_len function "
             "we can find the length of this chunk of text in tokens")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=400,
    chunk_overlap=20,
    length_function=tiktoken_len,
    separators=["\n\n", "\n", " ", ""]
)
```

```python
file1_path = '/content/iot_chat.txt'

# Read content of file 1
with open(file1_path, 'r') as file:
    iot_chat = file.read()
```

```python
import re

# Initialize dictionary to store messages
messages = {}

# Regular expression pattern to extract date and message
pattern = r'(\d{1,2}/\d{1,2}/\d{2}), (\d{1,2}:\d{2}) - (.+?)(?=\n\d{1,2}/\d{1,2}/\d{2}, \d{1,2}:\d{2}|$)'

# Find all matches using the pattern with the DOTALL flag
matches = re.findall(pattern, iot_chat, re.DOTALL)

# Iterate over matches and store in dictionary
for match in matches:
    date, time, content = match
    # Concatenate date and time to form datetime format if needed
    datetime = f"{date}, {time}"
    # Store in dictionary
    messages[datetime] = content.strip()

# Print the dictionary
for date, content in messages.items():
    print(f"Date: {date}")
    print(f"Content: {content}")
    print()
```

```python
from datetime import datetime

def parse_chat(chat):
    chunks = chat.strip().split('\n')
    chat_data = {}
    current_date = None
    current_messages = []

    for chunk in chunks:
        parts = chunk.split(' - ')
        if len(parts) == 2:
            date_str, message = parts
            try:
                # Parse date
                message_date = datetime.strptime(date_str, '%m/%d/%y, %H:%M')
                message_date_str = message_date.strftime('%Y-%m-%d')

                # If new date, store previous messages and update current date
                if current_date and current_date != message_date_str:
                    chat_data[current_date] = '\n'.join(current_messages)
                    current_messages = []
                current_date = message_date_str

                # Store message
                current_messages.append(message)
            except ValueError:
                # Skip lines that cannot be parsed as date-time
                pass

    # Store last batch of messages
    if current_date:
        chat_data[current_date] = '\n'.join(current_messages)

    return chat_data
```

```python
# Initialize a new dictionary to store filtered messages
filtered_messages = {}

# Iterate through the original dictionary and filter messages for year 2024
# Define a regular expression pattern for matching dates in the format */*/24*
pattern = re.compile(r'\d{1,2}/\d{1,2}/24')

# Filter out messages with dates in the year 2024 using regular expressions
filtered_messages = {date: content for date, content in messages.items() if pattern.match(date)}

# Print the filtered dictionary
for date, content in filtered_messages.items():
    print(f"Date: {date}")
    print(f"Content: {content}")
    print()
```

```python
filtered_dict = {}

for date_time, content in filtered_messages.items():
  date_only = date_time.split(',')[0]
  # Combine date with date-time and content
  new_content = f"Date Time: {date_time}\nPage Content: {content}"
  # Store in the new dictionary
  # Check if the date_only key already exists in the new_messages dictionary
  if date_only in filtered_dict:
      # If it exists, append the new content to the existing content
      filtered_dict[date_only] += '\n' + new_content
  else:
      # If it doesn't exist, set the new content as the value for the date_only key
      filtered_dict[date_only] = new_content
```

##Creating Embeddings

```python
import os

# get openai api key from platform.openai.com
OPENAI_API_KEY = os.getenv('OPENAI_API_KEY') or ''
```

```python
from langchain.embeddings.openai import OpenAIEmbeddings

model_name = 'text-embedding-ada-002'

embed = OpenAIEmbeddings(
    model=model_name,
    openai_api_key=OPENAI_API_KEY
)
```

### Vector Database

```python
from pinecone import Pinecone

# initialize connection to pinecone (get API key at app.pinecone.io)
api_key = os.getenv("PINECONE_API_KEY") or "YOUR_API_KEY"

# configure client
pc = Pinecone(api_key='')
```

```python
from pinecone import ServerlessSpec

spec = ServerlessSpec(
    cloud="aws", region="us-west-2"
)
```

```python
import time

index_name = 'rag-test'
existing_indexes = [
    index_info["name"] for index_info in pc.list_indexes()
]

# check if index already exists (it shouldn't if this is first time)
if index_name not in existing_indexes:
    # if does not exist, create index
    pc.create_index(
        index_name,
        dimension=1536,  # dimensionality of ada 002
        metric='dotproduct',
        spec=spec
    )
    # wait for index to be initialized
    while not pc.describe_index(index_name).status['ready']:
        time.sleep(1)

# connect to index
index = pc.Index(index_name)
time.sleep(1)
# view index stats
index.describe_index_stats()
```

```python
from langchain.vectorstores import Pinecone

text_field = "Chat Content"  # the metadata field that contains our text

# initialize the vector store object
vectorstore = Pinecone(
    index, embed.embed_query, text_field
)
```

```python
query = "Departmental magazine"

vectorstore.similarity_search(
    query,  # our search query
    k=3  # return 3 most relevant docs
)
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA

# completion llm
llm = ChatOpenAI(
    openai_api_key=OPENAI_API_KEY,
    model_name='gpt-3.5-turbo',
    temperature=0.0
)

qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vectorstore.as_retriever()
)
```
